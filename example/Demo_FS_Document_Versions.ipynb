```python
from ndi import Document
from ndi.database import FileSystem as Database, Query as Q
from test_app.subject_doc import Subject
```

```python
db = Database('../test_exp', document_extensions=[Subject])
```

```python
subject = Subject('reference','species', 'strain', 'variant')
```

```python
doc = Document('subject', subject, experiment_id='', file_id='', parent_id='')
```

```python
base_doc_id = doc.id
print(doc.id)
print(doc.version_depth)
print(doc.parent_id)
print(doc.asc_path)
```

```python
results = db.find(Document, Q('asc_path').contains(base_doc_id))
```

```python
for result in results:
    print(doc_ext := result.document_extension)
    print(f'asc_path: {result.asc_path}')
    print(f'parent_id: {result.parent_id}')
    print(f'document_id: {doc_ext.document_id}')
    print(f'reference: {doc_ext.reference}')
    print(f'variant: {doc_ext.variant}')
    print('')
```
