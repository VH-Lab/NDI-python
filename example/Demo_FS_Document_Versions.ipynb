```python
from ndi import Document, Query as Q
from ndi.database import FileSystem as Database
```

```python
# Connecting to database
db = Database('../test_exp')
```

```python
# Creating dummy document
doc = Document({ 'subject': 'mouse' })
doc.id
```

```python
doc.data['_metadata']
```

```python
# Adding document to database
db.add(doc)
```

```python
# Search document in database
result = db.find_by_id(doc.id)
print(result.data)
```

```python
doc.data['subject'] = 'elephant'
```

```python
# Save changed made to the document
doc.save_updates()
```

```python
# Search for document again to see that parent_id, version_depth, asc_path, and latest_version properties have been updated
result = db.find_by_id(doc.id)
print(result.data)
```

```python
doc.data['subject'] = 'elephants'
doc.save_updates()
```
