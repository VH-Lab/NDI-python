```python
from ndi import Experiment, DaqSystem, Probe
from ndi.database import FileSystem as Database, Query as Q
```

```python
db = Database('../test_exp')
```

```python
# Initialize empty experiment
e = Experiment('example')
```

```python
# The experiment is yet unbound to the database
e.ctx == None
# True
```

```python
# Add it to the database
db.add(e)
```

```python
# The experiment now has a context
e.ctx == db
# True
```

```python
# We could manually add a Probe
#   this method would set up the foreign keys on both objects (if applicable)
#   and add the probe to the database
e.add(Probe('probe params'))
```

```python
# More likely, we would set up a daq system to fill out the experiment for us
ds = DaqSystem('params..., file_navigator?')
```

```python
# This would be the step that really fills in the experiment with its Probes, Channels, etc.
#   ds.provision could take the experiment and add to it and the database
ds.provision(e)
```
