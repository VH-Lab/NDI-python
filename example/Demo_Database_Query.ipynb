```python
from ndi.database import Query as Q
```

```python
a = Q('a') == 'apple'
print(a)
print(a())
```

```python
b = Q('b') != 'bee'
print(b)
print(b())
```

```python
z = a & b
print(isinstance(z, Q))
print(z)
for query in z:
    print(query)
```

```python
y = a | b
print(isinstance(y, Q))
print(y)
for query in y:
    print(query)
```

```python
c = Q('c') >= 'cat'
print(c)
print(c())
```

```python
d = Q('d') < 'dog'
print(d)
print(d())
```

```python
for query in ((a | b) & ( c | d)):
    print(query)
```

```python
for query in ((a & b) | (c & d)):
    print(query)
```

```python
x = ((a & b) & c)
print(x)
for query in x:
    print(query)
```

```python
w = (a | b) | c
print(w)
for query in w:
    print(query)
```

```python
e = Q('e')[r'elephant']
print(e)
print(e())
```

```python
f = Q('f') / 'fish'
print(f)
print(f())
```
