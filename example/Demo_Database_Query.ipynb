```python
from ndi.database import Query as Q
```

```python
# a = Q('a').equals('apple')
a = Q('a') == 'apple'
print(a)
# print(a.query)
print(a())
```

```python
# a = Q('b').not_equals('bee')
b = Q('b') != 'bee'
print(b)
print(b())
```

```python
# z = a.and_(b)
# z = Q('a').equals('apple').and_(Q('b').not_equals('bee'))
# z = Q('a').equals('apple') & Q('b').not_equals('bee')
z = a & b
print(isinstance(z, Q))
print(z)
for query in z:
    print(query)
```

```python
# y = a.or_(b)
# y = Q('a').equals('apple').or_(Q('b').not_equals('bee'))
# y = Q('a').equals('apple') | Q('b').not_equals('bee')
y = a | b
print(isinstance(y, Q))
print(y)
for query in y:
    print(query)
```

```python
# c = Q('c').greater_than_or_equal_to(5)
c = Q('c') >= 5
print(c)
print(c())
```

```python
# d = Q('d').less_than(5)
d = Q('d') < 5
print(d)
print(d())
```

```python
# (a.or_(b)).and_((c.or_(d)))
for query in ((a | b) & ( c | d)):
    print(query)
```

```python
# (a.and_(b)).or_(c.and_(d))
for query in ((a & b) | (c & d)):
    print(query)
```

```python
x = ((a & b) & c)
print(x)
for query in x:
    print(query)
```

```python
w = (a | b) | c
print(w)
for query in w:
    print(query)
```

```python
e = Q('e').match(r'ele[a-z]$')
print(e)
print(e())
```

```python
f = Q('f').contains('fish')
print(f)
print(f())
```

```python
# g = Q('g').exists()
g = +Q('g')
print(g)
print(g())
```

```python
# h = Q('h').exists(False)
h = -Q('h')
print(h)
print(h())
```

```python
# i = Q('i').in_([1, 2, 3, 4])
i = Q('i') >> [1, 2, 3, 4]
print(i)
print(i())
```
