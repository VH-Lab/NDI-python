```python
from ndi.daqreaders.cedspike2 import CEDSpike2
```

```python
# Instantating DaqReader object with a data file
daqreader = CEDSpike2('../tests/data/daqreaders/cedspike2/example1.smr')
```

```python
# Acquiring sample rate from channel 1
samplerate = daqreader.samplerate(channel_number=1)
print(samplerate)
```

```python
# Reading data from channel 21 from index 0 to index 20
data = daqreader.readchannel(channel_number=21, start_index=0, end_index=20)
for i in data:
    print(i)
```

```python
# Reading event timestamps that are between 40 and 200 seconds
# Time units are in seconds
events = daqreader.readevents(channel_number=22, start_time=40, end_time=200)
for i in events:
    print(i)
```
