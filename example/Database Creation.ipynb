```python
from ndi import database
from ndi.database.types import Integer, String, Blob
```

```python
# Connect to database
db = database.Sql('postgres://postgres:1Password!@localhost:5432/practice')
```

```python
# Create table in database
Example = db.create_table('examples', {
    'name': String,
    'age': Integer,
    'data': Blob
})
```

```python
# Create object using schema defined in table creation
example1 = Example(name='Tommy', age=144, data=b'hello')
```

```python
# Add object to database
db.add(example1)
```

```python
# Search table in database based on query
results = db.find(Example, name='Tommy')
for result in results:
    print(result.id)
    print(result.name)
    print(result.age)
    print(result.data)
```

```python
results = db.find(Example, age=144)
for result in results:
    print(result.id)
    print(result.name)
    print(result.age)
    print(result.data)
```
