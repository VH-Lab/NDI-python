```python
from ndi import Document
from ndi.database import FileSystem as Database
```

```python
db = Database('../test_exp')
```

```python
doc = Document('noom', experiment_id='', file_id='', parent_id='')
```

```python
doc.id
```

```python
dependencies = [Document('naem', experiment_id='', file_id='', parent_id='') for _ in range(5)]
```

```python
for dependency in dependencies:
    db.add(dependency)
```

```python
dependency_ids = [dependency.id for dependency in dependencies]
doc.dependencies = dependency_ids
```

```python
doc.dependencies
```

```python
db.add(doc)
```

```python
db._collections['document_lookup'].find_dependencies(doc.id)
```

```python
db._collections['document_lookup'].find_dependents(doc.dependencies[0])
```

```python
result = db.find_by_id(Document, doc.id)
```
