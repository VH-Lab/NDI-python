```python
from ndi import Document
from ndi.database import FileSystem as Database
```

```python
db = Database('../test_exp')
```

```python
gas = Document({ 'PV': 'nRT' })
db.add(gas)
```

```python
force = Document({ 'F' : 'ma'})
force.add_dependency({'gas': gas.id })
db.add(force)
```

```python
doc = Document({'hello': 'world'})
doc.add_dependency({ 'force': force.id })
db.add(doc)
```
