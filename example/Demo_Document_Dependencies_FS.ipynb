```python
from ndi import Document
from ndi.database import FileSystem as Database
```

```python
db = Database('../test_exp')
```

```python
doc = Document(data={'hello': 'world'}, type_='test_doc', name='og', experiment_id='1234')
db.add(doc)
```

```python
force = Document({ 'F' : 'ma'}, 'equation', 'force', '1234')
doc.add_dependency(force)
```

```python
gas = Document({ 'PV': 'nRT' }, 'equation', 'gas', '1234')
force.add_dependency(gas)
```
