```python
from ndi.database import FileSystem as Database, Query as Q
from ndi import Probe
```

```python
db = Database('../test_exp')
```

```python
probe1 = Probe('Probe1', 1, 'ntrode')
probe2 = Probe('Probe2', 2, 'ntrode')
probe3 = Probe('Probe3', 3, 'sharp')
probe4 = Probe('Probe4', 4, 'sharp')
probe5 = Probe('Probe5', 5, 'patch')
probe6 = Probe('Probe6', 6, 'patch')
probe7 = Probe('Probe7', 4, 'stimulator')
probe8 = Probe('Probe8', 5, 'stimulator')
probe9 = Probe('Probe9', 6, 'stimulator')
```

```python
db.add([probe1, probe2, probe3, probe4, probe5, probe6, probe7, probe8, probe9])
```

```python
q = Q('name').contains('obe') & Q('type').match(r'.*a(t|o)')
q
```

```python
results = db.find(Probe, q)
for result in results:
    print(result.name)
```

```python
p = (Q('reference') > 4) | (Q('reference') <= 2)
p
```

```python
results = db.find(Probe, p)
for result in results:
    print(result.name)
```

```python
r = p & q
results = db.find(Probe, r)
for result in results:
    print(result.name)
```

```python
payload = {
    'name': 'fiodjwfodw'
}
db.update_many(Probe, r, payload)
results = db.find(Probe)
for result in results:
    print(result.name)
```

```python
s = Q('reference') > 3
db.delete_many(Probe, s)
results = db.find(Probe)
for result in results:
    print(result.name)
```
