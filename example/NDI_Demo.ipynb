```python
# Deprecated Example
from ndi import Channel, ChannelType, Probe, ProbeType, DaqSystem, Database, Experiment
import numpy as np
```

```python
# Creating Probe with channel data
c1 = Channel(0, ChannelType.digital_out, np.array([1., 2.]))
c2 = Channel(0, ChannelType.analog_in, np.array([3., 4.]))
p1 = Probe(name='P1', probe_type=ProbeType.ntrode, channels=[c1, c2])

# Creating Probe with channel data
c3 = Channel(0, ChannelType.auxiliary, np.array([5., 6.]))
c4 = Channel(0, ChannelType.time, np.array([7., 8.]))
p2 = Probe(name='P2', probe_type=ProbeType.electrode, channels=[c3, c4])

# Instantiating DaqSystem with Probes
ds1 = DaqSystem(name='DS1', probes=[p1, p2])
```

```python
# Creating Probe with channel data
c5 = Channel(0, ChannelType.mark, np.array([9., 10.]))
c6 = Channel(0, ChannelType.event, np.array([11., 12.]))
p3 = Probe(name='P3', probe_type=ProbeType.electrode, channels=[c5, c6])

# Creating Probe with channel data
c7 = Channel(0, ChannelType.analog_in, np.array([13., 14.]))
c8 = Channel(0, ChannelType.analog_in, np.array([15., 16.]))
p4 = Probe(name='P4', probe_type=ProbeType.brightfield_imaging, channels=[c7, c8])

# Instantiating DaqSystem with Probes
ds2 = DaqSystem(name='DS2', probes=[p3, p4])
```

```python
# Creating Databases
db1 = Database(path='DB1', experiment_unique_reference='hello')
db2 = Database(path='DB2', experiment_unique_reference='byebye')

# Instantiating Experiment with Databases and DaqSystems
exp = Experiment(reference='EXP',
                 databases=[db1, db2],
                 daq_systems=[ds1, ds2])
```

```python
# Creating Flatbuffer for Experiment
exp_buf = exp.serialize()

# Recreating Experiment from Flatbuffer
exp_2 = Experiment.frombuffer(exp_buf)
```

```python
# Testing new Experiment with original Experiment
assert exp == exp_2
assert exp.reference == exp_2.reference == 'EXP'
assert exp.databases[1].path == exp_2.databases[1].path == 'DB2'
assert exp.daq_systems[1].name == exp_2.daq_systems[1].name == 'DS2'
assert exp.daq_systems[0].probes[0].probe_type == exp_2.daq_systems[0].probes[0].probe_type == 'ntrode'
assert exp.daq_systems[0].probes[1].channels[0].data[1] == exp_2.daq_systems[0].probes[1].channels[0].data[1] == 6.

exp_2.reference = 'Exp'
assert exp != exp_2
assert exp.databases[1] == exp_2.databases[1]
```

```python
# Creating Flatbuffer for DaqSystem from within Experiment
ds2_buf = exp.daq_systems[1].serialize()

# Recreating DaqSystem from Flatbuffer
ds2_2 = DaqSystem.frombuffer(ds2_buf)
```

```python
# Testing new DaqSystem with DaqSystem in Experiment
assert exp.daq_systems[1] == ds2_2
assert exp.daq_systems[1].name == ds2_2.name == 'DS2'
assert exp.daq_systems[1].probes[0].channels[1].channel_type == ds2_2.probes[0].channels[1].channel_type == 'event'

ds2_2.name = 'ds2'
assert exp.daq_systems[1] != ds2_2
```
